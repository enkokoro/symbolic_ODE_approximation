```python
from approximation import ODE_approx
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numerical_methods import forward_euler_h
def f(t, x):
      return -x
def x_actual(x0, t):
      return x0*np.exp(-t)
M = 1
L = 2
T = 1
eps = 0.1
h = forward_euler_h(M, L, T, eps)
f_approx_fn = ODE_approx(f, M, L, T, eps=eps)
```

```python
x0 = 0.9
t0 = 0.5
print(h)
print(x_actual(x0, t0))
print(f_approx_fn(x0, t0))
```

```python
tt = np.linspace(0, T)
xx_actual = [x_actual(x0, t) for t in tt]
xx_approx = [f_approx_fn(x0, t) for t in tt]
plt.plot(tt, xx_actual, label="actual")
plt.plot(tt, xx_approx, label="approx")
plt.legend()
print(max(abs(np.array(xx_actual)-np.array(xx_approx))))
```
